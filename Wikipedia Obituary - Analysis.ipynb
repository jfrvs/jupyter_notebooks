## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib_inline
```

### Load deaths data, per decade and total

```python
df_90s = pd.read_csv("1990_-_2000.csv", index_col=0)
df_00s = pd.read_csv("2000_-_2010.csv", index_col=0)
df_10s = pd.read_csv("2010_-_2020.csv", index_col=0)
df_20s = pd.read_csv("2020_-_2024.csv", index_col=0)
df = pd.read_csv("1990_-_2024.csv", index_col=0)
```

### Load word counts, per decade and total

```python
wc_90s = pd.read_csv("1990_-_2000_-_Word_Count.csv", index_col=0)
wc_00s = pd.read_csv("2000_-_2010_-_Word_Count.csv", index_col=0)
wc_10s = pd.read_csv("2010_-_2020_-_Word_Count.csv", index_col=0)
wc_20s = pd.read_csv("2020_-_2024_-_Word_Count.csv", index_col=0)
wc = pd.read_csv("1990_-_2024_-_Word_Count.csv", index_col=0)
```

### Define list of words to remove from counts

```python
removed_words = ("and", "of", "the", "in", "new", "by", "ii", "a", "from", "for", "to", "at", "after", "us", "on")
```

### Clean word counts, choose top 100 words

```python
wc_90s = wc_90s[~wc_90s["Word"].isin(removed_words)].head(100)
wc_00s = wc_00s[~wc_00s["Word"].isin(removed_words)].head(100)
wc_10s = wc_10s[~wc_10s["Word"].isin(removed_words)].head(100)
wc_20s = wc_20s[~wc_20s["Word"].isin(removed_words)].head(100)
wc = wc[~wc["Word"].isin(removed_words)].head(100)
```

### Word count charts, with top 15 words

```python
plt.figure().set_figwidth(15)
plt.bar(wc_90s["Word"].head(15), wc_90s["Count"].head(15))
plt.show()
```

```python
plt.figure().set_figwidth(15)
plt.bar(wc_00s["Word"].head(15), wc_00s["Count"].head(15))
plt.show()
```

```python
plt.figure().set_figwidth(15)
plt.bar(wc_10s["Word"].head(15), wc_10s["Count"].head(15))
plt.show()
```

```python
plt.figure().set_figwidth(15)
plt.bar(wc_20s["Word"].head(15), wc_20s["Count"].head(15))
plt.show()
```

```python
plt.figure().set_figwidth(15)
plt.bar(wc["Word"].head(15), wc["Count"].head(15))
plt.ylabel("Count")
plt.box(False)
plt.show()
```

1. When examining the top 15 words from 1990 to 2023, we have 5 nationalities, 8 occupations and 2 words that make reference to causes of death, cancer and heart disease
2. The mention of indian individuals is more prominent in the last 15 years, but are sufficient to be placed in the top 15 from 1990 to 2023
3. Americans outnumber the second placed nationality by more than 3 to 1
4. Analysis of the 20s revealed the significance of the covid pandemic, with multiple victims associated with the disease
5. Besides that, the trends were largely the same, with some minor variations

### Age analysis

```python
print("Mean age of death in the 90s: {:.2f}".format(df_90s['Age'].mean()))
print("Mean age of death in the 00s: {:.2f}".format(df_00s['Age'].mean()))
print("Mean age of death in the 10s: {:.2f}".format(df_10s['Age'].mean()))
print("Mean age of death in the 20s: {:.2f}".format(df_20s['Age'].mean()))
print("Mean age of death overall: {:.2f}".format(df['Age'].mean()))
```

```python
plt.subplot(221)
df_90s['Age'].plot(kind='hist', bins = 12)
plt.xlim(0, 120)
plt.yticks([])
plt.title("1990s", x=0.2, y=0.8)

plt.subplot(222)
df_00s['Age'].plot(kind='hist', bins = 12)
plt.xlim(0, 120)
plt.yticks([])
plt.title("2000s", x=0.2, y=0.8)

plt.subplot(223)
df_10s['Age'].plot(kind='hist', bins = 12)
plt.xlim(0, 120)
plt.yticks([])
plt.xlabel("Age")
plt.title("2010s", x=0.2, y=0.8)

plt.subplot(224)
df_20s['Age'].plot(kind='hist', bins = 12)
plt.xlim(0, 120)
plt.yticks([])
plt.xlabel("Age")
plt.title("2020s", x=0.2, y=0.8)
plt.show()
```
